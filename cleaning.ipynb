```python
import pandas as pd
import numpy as np
import re

df = pd.read_json(r"/exports/zonaprop_data_total.json")

display(df.head())
```

```python
def clean_text(text):
    # Remove leading and trailing whitespace, newlines, and tabs
    cleaned = text.strip()
    # Replace multiple spaces, tabs, and newlines with a single space
    cleaned = re.sub(r'[\t\n\r]+', ' ', cleaned)
    return cleaned

def extract_number(text):
    # Check if the input is a list and join its elements into a single string
    if isinstance(text, list):
        text = " ".join(text)
    # Use regex to extract numbers with possible periods or commas
    # Replace commas with empty string and handle decimals correctly
    text = text.replace(",", "").replace(".", "")
    # Find numbers in the string
    numbers = re.findall(r'\d+', text)
    # Return the first number as an integer
    return int(numbers[0]) if numbers else None



def join_clean_text(text_list):
    # Join the list elements into one string and strip excess spaces
    joined_text = " ".join(text_list)
    # Remove extra spaces, newlines, and tabs
    cleaned_text = " ".join(joined_text.split())
    return cleaned_text



def split_address(row):
    # Separar la dirección en partes
    address_split = row['direccion_barrio_ciudad'].split(',')
    address_split = [item.strip() for item in address_split]  # Limpiar espacios

    # Inicializar variables
    direccion = barrio = ciudad = None

    # Asignar valores según la cantidad de elementos
    if len(address_split) == 3:
        direccion, barrio, ciudad = address_split
    elif len(address_split) == 2:
        barrio, ciudad = address_split
        direccion = ""

    return pd.Series([direccion, barrio, ciudad])


df['metros_totales_clean'] = df['metros_totales'].apply(extract_number)
df['descripcion_clean'] = df['descripcion'].apply(join_clean_text)
df[['direccion', 'barrio', 'ciudad']] = df.apply(split_address, axis=1)
df['precio_cleaned'] = df['precio'].apply(extract_number)
# Handle NaN values by filling them with a specific value, e.g., 0, or drop them
df['precio_cleaned'] = df['precio_cleaned'].fillna(0).astype(int)  # Replace NaNs with 0 and convert to int
print(df.dtypes)
display(df.head())
```

```python
df.to_excel(r"/exports/zonaprop_cleaned.xlsx", index=False)
```
